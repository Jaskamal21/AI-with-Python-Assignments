```python
import numpy as np
import matplotlib.pyplot as plt

n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

for n in n_values:
    dice1 = np.random.randint(1, 7, n)
    dice2 = np.random.randint(1, 7, n)
    s = dice1 + dice2  

    h, h2 = np.histogram(s, bins=range(2, 14))
    
    plt.figure(figsize=(8, 5))
    plt.bar(h2[:-1], h / n, color='blue', alpha=0.7)
    plt.xlabel("Sum of Two Dice")
    plt.ylabel("Frequency")
    plt.title(f"Histogram of Dice Sums (n={n})")
    plt.xticks(range(2, 13))
    plt.show()
```

```python
5.) What do you observe? You may need to run the loop a few times to see it.
Observation: When n increases, the histogram becomes smoother and more closely resembles the theoretical probabilities:
For small values of n (e.g., 500 or 1000), the histogram might be irregular, as random fluctuations are more noticeable with fewer samples.
For larger values of n (e.g., 50,000 or 100,000), the histogram closely approximates the expected probabilities, with a clear peak around 7
and tapering off towards 2 and 12.

6.)How is this related to "regression to the mean"?
This experiment demonstrates regression to the mean, where larger sample sizes tend to average out fluctuations, leading the
observed frequencies to converge to the theoretical probabilities. In this case, as n increases, the distribution of sums "regresses" towards the
expected distribution, showing the central tendency more clearly.
```
